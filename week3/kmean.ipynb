```python
#import the libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# Generate syntatic data using make_blods

samples, true_labels = make_blobs(n_samples=500,centers=5,cluster_std=0.85, random_state=0)
```

```python
# printing five elements for each of the variables returned
print(samples.shape)
print("Samples of muy data points: \n",samples[:5])
print("True lable or category: ",true_labels[:5])
```

```python
print(true_labels)
```

```python
#plotting dataset
plt.scatter(samples[:,0],samples[:,1],s=20)
plt.show()
```

```python
# running the Kmean model
kmeans = KMeans(init="random", n_clusters=5, n_init=10, max_iter=300, random_state=0)
kmeans.fit(samples)
y_kmeans = kmeans.predict(samples)
```

```python
plt.scatter(samples[:,0], samples[:,1],c=y_kmeans, cmap='spring')
center = kmeans.cluster_centers_
plt.scatter(center[:,0],center[:,1],c='blue', s=60, alpha=0.9)
plt.show()
```

```python
# The clusters assignments are store into 1-D array.
print(kmeans.labels_[:5])
print(true_labels[:5])
print(kmeans.labels_)
```

```python
# Running the model using different configuration.
kmean_kwarg={"init":"random","n_init":10, "max_iter":1000, "random_state":42}

maximum_number_cluster = 10
sse = []
for k in range(1,maximum_number_cluster):
    kmeans = KMeans(n_clusters=k, **kmean_kwarg)
    kmeans.fit(samples)
    sse.append(kmeans.inertia_)


```

```python
plt.plot(range(1,maximum_number_cluster),sse)
plt.xticks(range(1,maximum_number_cluster))
plt.xlabel("Number of Cluster")
plt.ylabel("SSE")
```
